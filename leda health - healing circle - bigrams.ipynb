```python

#What was the most helpful part of the session?
data_1 = pd.read_excel("/Users/macbookretina/Desktop/Session_Exit_Survey2021-07-22_14_51_18.xlsx", 
                       usecols = [0,4], parse_dates=True, 
                       names=['Submission Date', 'Helpful'],
                       dtype={'Submission Date': str, 'Helpful': str})

#mood_data_1 = data_1.sort_values('Mood Ratings')
mood_data_1 = data_1
#mood_data_1 = mood_data_1.dropna()
mood_data_1["Group"] = "Big Circle"
mood_data_1
```

```python
#art circle

data_2 = pd.read_excel("/Users/macbookretina/Desktop/Session Exit Survey - Art Circle (Responses).xlsx", 
                       usecols = [0,4], parse_dates=True, 
                       names=['Submission Date', 'Helpful'],
                       dtype={'Submission Date': str, 'Helpful': str})

#mood_data_2 = data_2.sort_values('Mood Ratings')
mood_data_2 = data_2
#mood_data_1 = mood_data_1.dropna()
mood_data_2["Group"] = "Art Circle"
mood_data_2
```

```python
#queer circle

data_3 = pd.read_excel("/Users/macbookretina/Desktop/Exit Survey - Queer Circle (Responses).xlsx", 
                       usecols = [0,4], parse_dates=True, 
                       names=['Submission Date', 'Helpful'],
                       dtype={'Submission Date': str, 'Helpful': str})

#mood_data_3 = data_3.sort_values('Mood Ratings')
mood_data_3 = data_3
#mood_data_1 = mood_data_1.dropna()
mood_data_3["Group"] = "Queer Circle"
mood_data_3
```

```python
#append improvement

data_imp_1 = mood_data_1.append(mood_data_2)

data_imp = data_imp_1.append(mood_data_3)

data_imp.info()
```

```python
data_imp["index"] = 0

for i in range(143):
    data_imp.iloc[i, data_imp.columns.get_loc('index')] = i
    
    
data_imp = data_imp.set_index('index')
data_imp
```

```python
group_data_2 = data_imp.sort_values('Group')
group_data_2 = group_data_2.dropna()
group_data_2
```

```python
group_data_2['ca_id'] = group_data_2['Group'].factorize()[0]
category_id_df = group_data_2[['Group', 'ca_id']].drop_duplicates().sort_values('ca_id')
category_to_id = dict(category_id_df.values)
id_to_category = dict(category_id_df[['ca_id', 'Group']].values)
```

```python
group_data_2
```

```python
# FIND THE MOST USED UNIGRAMS AND BIGRAMGS
singles = ""
#doubles = ""

model=LinearSVC()

# GET THE TERM FREQUENCY AND INVERSE DOCUEMNT FREQUENCY scores of the words all at once
tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 2), stop_words='english')
features = tfidf.fit_transform(group_data_2.Helpful).toarray()
labels = group_data_2.ca_id

model.fit(features, labels)
N = 2
for industry, in_id in sorted(category_to_id.items()):
  indices = np.argsort(model.coef_[in_id])
  feature_names = np.array(tfidf.get_feature_names())[indices]
  unigrams = [v for v in reversed(feature_names) if len(v.split(' ')) == 1][:N]
#  bigrams = [v for v in reversed(feature_names) if len(v.split(' ')) == 2][:N]

  for i in range(2):
      singles = singles + " " + unigrams[i]
#      doubles =  doubles + " " + bigrams[i]
  print("# '{}':".format(industry))
  print("  . Top unigrams:\n       . {}".format('\n       . '.join(unigrams)))
#  print("  . Top bigrams:\n       . {}".format('\n       . '.join(bigrams)))
```
